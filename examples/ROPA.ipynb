```python
import os
import sys
pp_path = os.path.join("..")
sys.path.append(pp_path)
from pySMOKEpostprocessor import *
import matplotlib.pyplot as plt
```

## Rate Of Production Analysis

```python
kineticFolder = os.path.join("ROPA", "kinetics")
resultsFolder = os.path.join("ROPA", "Output")

PostProcessor = pySMOKEpostproccesor(kineticmechanismFolder = kineticFolder, 
                                     simulationresultsfolder = resultsFolder, 
                                     verbose=False)
```

```python
specie = "H2"

coefficients, indexes = PostProcessor.RateOfProductionAnalysis(specie=specie, 
                                                               ropatype='global', 
                                                               numberofreactions=15)
```

```python
colori = []
names = [f"R{i}" for i in indexes]
for i in coefficients:
    if (i>=0):
        colori.append('red')
    else:
        colori.append('blue')
plt.rcParams["figure.figsize"]=18,15
plt.title("$H_{2}$ Global Rate Of Production Analysis", fontsize=22)
plt.barh(names,coefficients, color = colori)
plt.yticks(fontsize = 20)
plt.xticks(fontsize = 20)
plt.gca().invert_yaxis()
plt.show()
```
