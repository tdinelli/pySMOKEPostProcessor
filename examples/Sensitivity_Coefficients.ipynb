```python
import os
import pySMOKEPostProcessor as pp
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

**SensitivityCoefficients**: function that returns the Sensitivity coefficient of a specific target for a specific reaction.
   > * `target` ( *string* ): name of the target for which is desired the plot of the sensitivity coefficient please take into consideration that is necessary to run the simulation for the target in order to post-process it.
   > * `reaction_name` ( *string* ): name of the reaction to be analyzed.
   > * `normalization_type` ( *string* ): sets the type of normalization that sensitivity coefficients undergoes inside the post-processing. Available types are: "local" and "max_value".
   1. **local**: the sensitivity coefficients are normalized using the local value of the Y variable (specie). This approach can lead to unspected results when the Y variable approaches 0 (i.e., when the mass fraction of a specie is close to zero).
   2. **max_value**: the sensitivity coefficients are normalized using the maximum value of the Y variable (specie) evaluated on the whole computational domain. 

```python
kineticFolder = os.path.join("data", "Sensitivity-01", "kinetics")
resultsFolder = os.path.join("data", "Sensitivity-01", "Output-01")
time, sensitivity_coefficient = pp.SensitivityCoefficients(kinetic_folder=kineticFolder,
                                                        output_folder=resultsFolder,
                                                        target='O2', 
                                                        reaction_name='O2+H=O+OH', 
                                                        normalization_type='max-value')

plt.plot(time, sensitivity_coefficient, label = 'R1: O2+H=O+OH')
plt.xlabel('time [s]', fontsize=20)
plt.ylabel('$S^{O_{2}}_{R_{1}}$', fontsize=20)
plt.legend()
plt.grid()
plt.show()
```
