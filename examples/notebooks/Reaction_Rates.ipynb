```python
import os
import matplotlib.pyplot as plt
from pySMOKEPostProcessor import PostProcessor
from pySMOKEPostProcessor import OpenSMOKEppXMLFile
```

### GetReactionRates
function that returns the reaction rate in [kmol/m3/s] of a specific reaction.
   > * `abscissae_name` ( *string* ): name of the abscissae for which is desired the plot of the reaction rate
   > * `reaction_name` (list(*string*)): names of the reactions for which is required the reaction rate (not read if reaction_index is specified)
   > * `reaction_index` (list(*int*)): indices of the reactions (not read if reaction_name is specified)
   > * `sum_rates` ( *bool* ): sum the rates of the reactions from the list provided (default: false)

```python
kineticFolder = os.path.join("..", "data", "ROPA", "kinetics")
resultsFolder = os.path.join("..", "data", "ROPA", "Output")

pp = PostProcessor(kineticFolder, resultsFolder)
output = OpenSMOKEppXMLFile(resultsFolder, kineticFolder)

reaction_rate = pp.GetReactionRates(reaction_name=['O2+H=O+OH'])[0]
plt.plot(output.T, reaction_rate, 'r')
plt.xlabel('Temperature [K]', fontsize=15)
plt.ylabel('Reaction Rate [kmol/m3/s]', fontsize=15)
plt.grid()
plt.show()
```

```python
reaction_rate_1 = pp.GetReactionRates(reaction_name = ['O2+H=O+OH'])[0]

reaction_rate_2 = pp.GetReactionRates(reaction_name = ['H2+O=H+OH'])[0]

reaction_rate_3 = pp.GetReactionRates(reaction_name = ['O2+H(+M)=HO2(+M)'])[0]

plt.plot(output.T, reaction_rate_1, 'r', label = 'O2+H=O+OH')
plt.plot(output.T, reaction_rate_2, 'b', label = 'H2+O=H+OH')
plt.plot(output.T, reaction_rate_3, 'g', label = 'O2+H(+M)=HO2(+M)')
plt.xlabel('Temperature [K]', fontsize=15)
plt.ylabel('Reaction Rate [kmol/m3/s]', fontsize=15)
plt.legend()
plt.grid()
plt.show()
```

```python
# as above, but sum the reaction rates 
reaction_rates = pp.GetReactionRates(reaction_name = ['O2+H=O+OH', 'H2+O=H+OH', 'O2+H(+M)=HO2(+M)'], 
                                     sum_rates = True)[0]
plt.plot(output.T, reaction_rates, 'r', label = 'summed rates')
plt.xlabel('Temperature [K]', fontsize=15)
plt.ylabel('Reaction Rate [kmol/m3/s]', fontsize=15)
plt.legend()
plt.grid()
plt.show()
```
